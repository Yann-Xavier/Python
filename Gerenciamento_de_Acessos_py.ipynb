```python
class Autorizacao:
    def __init__(self):
        self.funcionarios = set()  # Conjunto para armazenar funcionários autorizados
        self.visitantes = set()    # Conjunto para armazenar visitantes autorizados

    def exibir_menu(self):
        # Exibir menu
        print("\nMENU:")
        print("1. Adicionar funcionário autorizado")
        print("2. Adicionar visitante autorizado")
        print("3. Verificar autorização de acesso")
        print("4. Remover autorização de acesso")
        print("5. Consultar listas de autorizados")
        print("6. Sair")

    def adicionar_pessoa(self, conjunto, tipo):
        nome = input(f"Digite o nome do {tipo}: ").strip().capitalize() # Trata espaços em excesso e capitaliza o nome
        if nome:
            if nome in conjunto:
                print(f"O {tipo} {nome} já está autorizado.")
            else:
                conjunto.add(nome)
                print(f"{tipo.capitalize()} {nome} adicionado com sucesso.")
        else:
            print("Nome inválido. Por favor, insira um nome válido.")

    def verificar_autorizacao(self):
        nome = input("Digite o nome da pessoa: ").strip().capitalize() # Trata espaços em excesso e capitaliza o nome
        if nome:
            if nome in self.funcionarios:
                print(f"{nome} está autorizado a acessar o prédio como funcionário.")
            elif nome in self.visitantes:
                print(f"{nome} está autorizado a acessar o prédio como visitante.")
            else:
                print(f"{nome} não está autorizado a acessar o prédio.")
        else:
            print("Nome inválido. Por favor, insira um nome válido.")

    def remover_autorizacao(self):
        nome = input("Digite o nome da pessoa a ser removida: ").strip().capitalize() # Trata espaços em excesso e capitaliza o nome
        if nome:
            if nome in self.funcionarios:
                self.funcionarios.remove(nome)
                print(f"Funcionário {nome} removido com sucesso.")
            elif nome in self.visitantes:
                self.visitantes.remove(nome)
                print(f"Visitante {nome} removido com sucesso.")
            else:
                print(f"{nome} não encontrado.")
        else:
            print("Nome inválido. Por favor, insira um nome válido.")

    def consultar_listas(self):
        print("\nFuncionários autorizados:")
        for funcionario in sorted(self.funcionarios):
            print(funcionario)
        print("\nVisitantes autorizados:")
        for visitante in sorted(self.visitantes):
            print(visitante)

    def main(self):
        while True:
            self.exibir_menu()
            opcao = input("Escolha uma opção: ")

            # Verificar opção escolhida
            if opcao == '1':
                self.adicionar_pessoa(self.funcionarios, "funcionário")
            elif opcao == '2':
                self.adicionar_pessoa(self.visitantes, "visitante")
            elif opcao == '3':
                self.verificar_autorizacao()
            elif opcao == '4':
                self.remover_autorizacao()
            elif opcao == '5':
                self.consultar_listas()
            elif opcao == '6':
                print("Encerrando o programa...")
                break
            else:
                print("Opção inválida! Por favor, escolha uma opção válida.")

if __name__ == "__main__":
    Autorizacao().main()
```
