```python
def adicionar_produto(estoque):
    # Função para adicionar um produto ao estoque
    nome_produto = input("Digite o nome do produto: ")
    quantidade = int(input("Digite a quantidade a ser adicionada: "))
    estoque[nome_produto] = estoque.get(nome_produto, 0) + quantidade
    print("Produto adicionado com sucesso!")

def atualizar_quantidade(estoque):
    # Função para atualizar a quantidade de um produto no estoque
    nome_produto = input("Digite o nome do produto: ")
    nova_quantidade = int(input("Digite a nova quantidade: "))
    estoque[nome_produto] = nova_quantidade
    print("Quantidade atualizada com sucesso!")

def remover_produto(estoque):
    # Função para remover um produto do estoque
    nome_produto = input("Digite o nome do produto a ser removido: ")
    if nome_produto in estoque:
        del estoque[nome_produto]
        print("Produto removido com sucesso!")
    else:
        print("Produto não encontrado no estoque.")

def exibir_estoque(estoque):
    # Função para exibir o estoque atual
    print("Estoque atual:")
    for produto, quantidade in estoque.items():
        print(f"{produto}: {quantidade}")

def main():
    # Função principal do programa
    estoque = {}  # Inicializa o estoque como um dicionário vazio
    continuar = True  # Variável para controlar a execução do loop while

    while continuar:
        # Exibe o menu de opções para o usuário
        print("\nMenu:")
        print("1. Adicionar produto")
        print("2. Atualizar quantidade de um produto")
        print("3. Remover produto")
        print("4. Exibir estoque")
        print("5. Sair do programa")

        try:
            # Solicita ao usuário que escolha uma opção do menu
            opcao = int(input("Escolha uma opção: "))

            if opcao == 1:
                adicionar_produto(estoque)
            elif opcao == 2:
                atualizar_quantidade(estoque)
            elif opcao == 3:
                remover_produto(estoque)
            elif opcao == 4:
                exibir_estoque(estoque)
            elif opcao == 5:
                # Encerra o loop while e finaliza o programa
                continuar = False
                print("Saindo do programa...")
            else:
                # Mensagem de erro para opção inválida
                print("Opção inválida. Escolha um número de 1 a 5.")
        except ValueError:
            # Trata exceção caso o usuário digite um valor inválido (não inteiro)
            print("Por favor, digite um número válido.")

if __name__ == "__main__":
    main()
```
